```python
from meteostat import Point, Daily
from datetime import datetime
import pandas as pd

# Coordenadas de Bocas del Toro, Panamá
bocas_del_toro = Point(9.3400, -82.2400, 11)

# Rango de fechas: año 2024
start = datetime(2024, 1, 1)
end = datetime(2024, 12, 31)

# Obtener los datos diarios
data = Daily(bocas_del_toro, start, end)
data = data.fetch()

# Guardar directamente como JSON en la ruta deseada
data.reset_index().to_json("../../data/raw/datos_bocas_del_toro_crudos.json", orient="records", date_format="iso")
```
